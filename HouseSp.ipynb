```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Housing.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
binary_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for col in binary_columns:
    df[col] = df[col].map({'yes': 1, 'no': 0})
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['furnishingstatus'], drop_first=False)
```

```python
df
```

```python
df[["furnishingstatus_furnished","furnishingstatus_semi-furnished", "furnishingstatus_unfurnished"]] = df[["furnishingstatus_furnished","furnishingstatus_semi-furnished", "furnishingstatus_unfurnished"]].astype(int)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

# taking features or columns that were having quantity data 

```python
numerical_columns = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']
```

```python
scaler = StandardScaler()
```

```python
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
df
```

```python
X = df.drop('price', axis=1) 
```

```python
X
```

```python
Y = df['price']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state= 500)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
